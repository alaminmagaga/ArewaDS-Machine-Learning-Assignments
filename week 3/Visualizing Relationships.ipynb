```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('honey.csv')
df.head()
```

```python
df.info()
```

```python
sns.pairplot(df)
```

```python
sns.relplot(data=df,
            x='year',y='stocks',col='state',col_wrap=3,kind='line')
```

```python
fig, ax = plt.subplots(figsize=(12,6))
lineplot = sns.lineplot(x=df['year'], y=df['numcol'], data=df, 
                        label = 'Number of bee colonies', legend=False)
sns.despine()
plt.ylabel('# colonies')
plt.title('Price vs number of bee colonies over the years');

ax2 = ax.twinx()
lineplot2 = sns.lineplot(x=df['year'], y=df['priceperlb'], ax=ax2, color="r", 
                         label ='Price per lb', legend=False) 
sns.despine(right=False)
plt.ylabel('price per lb')
ax.figure.legend()
```

```python
sns.relplot(x="priceperlb", y="state", size="year", data=df, height=15, aspect=.5)
```

```python
sns.relplot(x="year", y="totalprod", kind="line", data=df)
```
